# Importando os dados

Importando as bibliotecas que serão utilizadas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Carregando os dados, tratando o cabeçalho e validando se todas as informações entraram corretamente.

```python
#Link dos dados (planilha com informações de preços dos carros)
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
#Novos nomes das colunas
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
#Lendo o arquivo CSV e atualizando os headers
df = pd.read_csv(url, names=headers)
#Mostrando as 5 primeiras linhas do arquivo
df.head()
```

# Encontrando os dados nulos.

Substituindo os valores marcados como "?" (Vazios) por NaN

```python
df.replace("?", np.nan, inplace = True)
df.head()
```

Avaliando os valores nulos no dataframe.

```python
dados_faltantes = df.isnull()
dados_faltantes.head()
```

Contando os valores faltantes em cada coluna do dataframe

```python
for column in dados_faltantes.columns.tolist():
  print(column)
  print(dados_faltantes[column].value_counts())
  print("")
```

Calculando e sustituindo os valores das colunas onde a solução atribuida foi substituir pela media de valores.

```python
media_norm_loss = df["normalized-losses"].astype(float).mean(axis=0)
print("Media das perdas normalizadas: ", media_norm_loss)
media_stroke = df["stroke"].astype(float).mean(axis=0)
print("Media das stroke: ", media_stroke)
media_bore = df["bore"].astype(float).mean(axis=0)
print("Media das bore: ", media_bore)
media_horsepower = df["horsepower"].astype(float).mean(axis=0)
print("Media dos cavalos de potencia: ", media_horsepower)
media_peakrpm = df["peak-rpm"].astype(float).mean(axis=0)
print("Media dos cavalos de potencia: ", media_peakrpm)

df["normalized-losses"].replace(np.nan, media_norm_loss, inplace=True)
df["stroke"].replace(np.nan, media_stroke, inplace=True)
df["bore"].replace(np.nan, media_bore, inplace=True)
df["horsepower"].replace(np.nan, media_horsepower, inplace=True)
df["peak-rpm"].replace(np.nan, media_peakrpm, inplace=True)
```

Substituindo os valores onde a solução encontrada foi substituir os valores pela frequencia.

```python
#Retorna a quantidade de valores
print(df['num-of-doors'].value_counts())
#Retorna o valor com maior quantidade
df['num-of-doors'].value_counts().idxmax()
#Substitui os valores das portas pelo valor de maior frequencia
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

Excluindo os dados em que a solucação encontrada foi excluir as linhas dos dados.

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

# Corrigindo o formato dos dados

Primeiramente antes de corrigirmos os tipos de dados precisamos verificar qual o tipo de cada coluna.

```python
#Verificando os dados de todas as colunas
df.dtypes
```

Após isso podemos definir o melhor tipo de dado para cada coluna com base nas informações de dados que ja temos

```python

#Convertendo os dados para o formato apropriado
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
#Validando os formatos dos dados após a atualização
df.dtypes
```
